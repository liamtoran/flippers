```python
def get_L(n=100, n_weak=20):
    L = np.round(0.25 * np.random.rand(n, n_weak), 2)
    L[:, :5] = 4 * L[:, :5]
    L[0] = 0
    L = pd.DataFrame(L, columns=[f"lf_{i}" for i in range(n_weak)])
    return L
```
